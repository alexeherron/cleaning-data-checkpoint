```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
from wordcloud import WordCloud, STOPWORDS
```

```python
coffee_words = pd.read_csv('clean_coffee.csv', index_col=False)
coffee_words.head()
```

```python
coffee_words = coffee_words.drop('Unnamed: 0', axis=1)
coffee_words.head()
```

```python
coffee_words['blind_assessment'] = coffee_words['blind_assessment'].str.replace(',', '').str.lower()
coffee_words.head()
```

```python
coffee_words['blind_assessment'] = coffee_words['blind_assessment'].str.replace('.', '')
coffee_words['blind_assessment'] = coffee_words['blind_assessment'].str.split(' ')
coffee_words.head()
```

```python
print(coffee_words['blind_assessment'][0])
```

```python
coffee_words.to_csv('coffee_words.csv')
```
